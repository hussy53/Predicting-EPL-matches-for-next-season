```python
# Clean up data
# Machine Learning libraries work with numeric data and not objects
matches.dtypes
```

```python
# Convert date from string to time - easier to work with ML
matches['date'] = pd.to_datetime(matches['date'])
matches.dtypes
```

```python
# 0 - Away
# 1 - Home
matches['venue_code'] = matches['venue'].astype('category').cat.codes
matches
```

```python
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
matches['hour'] = matches['time'].str.replace(':.+', '', regex=True).astype('int')
matches['day_code'] = matches['date'].dt.dayofweek
matches['target'] = (matches['result'] == 'W').astype('int')
```

```python
matches
```

```python
# RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches['date'] < '2022-01-01']
test = matches[matches['date'] > '2022-01-01']
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
# Fit the model - train the predictors, trying to predict 'target'
rf.fit(train[predictors], train['target'])
```

```python
# Generate the predictions
preds = rf.predict(test[predictors])
```

```python
# How accurate was the prediction
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'], preds)
# 61% accurate
acc
```

```python
combined = pd.DataFrame(dict(actual=test['target'], prediction=preds))
```

```python
pd.crosstab(index=combined['actual'], columns=combined['prediction'])
# When we predicted a 'L', we were right (146%) of the times compared to predicting wrong (83%)
# But we were wrong more often (26%) compared to predicting right (21%) when predicting a 'W'
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test['target'], preds)
# Checking our precision - we were 45% right when predicting a 'W' - not good precision
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")# Group the dates from lowest to highest
    rolling_stats = group[cols].rolling(3).mean()# Take the first 3 match results, exclude the current
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)# drop the null values(missing data)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f'{c}_rolling' for c in cols]# f for function(c)
```

```python
rolling_averages(group, cols, new_cols)
```

```python
# Group each team and apply averages to those teams
matches_rolling = matches.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
# Give new indices to each row
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
# Retaining the ML model - making new set of predictions
def make_predictions(data, predictors):
    # Train/Test data
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] > '2022-01-01']
    # Fit the model - train the predictors, trying to predict 'target'
    rf.fit(train[predictors], train['target'])
    # Generate the predictions
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], predicted=preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
# Improved precision - 70%
precision
```

```python
# Matches index in matches_rolling and merge the specified rows to see which data is missing that caused a difference in precision
combined = combined.merge(matches_rolling[['date', 'team', 'opponent', 'result']], left_index=True, right_index=True)
combined
```

```python
# Deal with inconsistent team names
class InconsistentDict(dict):
    __missing__ = lambda self, key : key

map_values = {
    "Brighton and Hove Albion" : "Brighton",
    "Manchester United" : "Manchester Utd",
    "Newcastle United" : "Newcastle Utd",
    "Tottenham Hotspur" : "Tottenham",
    "West Ham United" : "West Ham",
    "Wolverhampton Wanderers" : "Wolves"
}
mapping = InconsistentDict(**map_values)
```
